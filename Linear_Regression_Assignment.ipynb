```python
import numpy as np
import matplotlib.pyplot as plt

# This is for generating random data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.rand(100, 1)

# Process for Linear Regression using Numpy as a library
X_b = np.c_[np.ones((100, 1)), X]  # Add bias term (intercept)
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# Predict the data using the obtained parameters
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_best)

# Plot the data points and regression line
plt.scatter(X, y, alpha=0.6, label='Data Points')
plt.plot(X_new, y_predict, 'r-', label='Linear Regression Line', linewidth=2)
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.title('Linear Regression For Randomized Generated Data')
plt.show()
```
